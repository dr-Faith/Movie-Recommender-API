```python
def clean_year(year_raw):
    return year_raw.split('â€“')[0]

def clean_director(director_raw):
    return director_raw.split('|')[0].strip()

def clean_rating(rating_raw):
    if rating_raw == 'None':
        return 0
    else:
        return rating_raw
    
def clean_votes(votes_raw):
    if votes_raw == 'None':
        return 0
    else:
        return votes_raw
    
def change_dtypes(df):
    df['year'] = df['year'].astype('category')
    df['duration'] = df['duration'].astype(np.int64)
    df['rating'] = df['rating'].astype(float)
    df['votes'] = df['votes'].str.replace(',', '').apply(lambda x: np.int(x) if pd.notnull(x) else 0)
    
def clean_data(df, save_name):
    df.drop_duplicates(inplace=True)
    df = df[(df['movie']!='None') & (df['genres']!='None') & (df['rating']!='None')]
    df['year'] = df['year'].apply(clean_year)
    df['director'] = df['director'].apply(clean_director)
    df['rating'] = df['rating'].apply(clean_rating)
    df['votes'] = df['votes'].apply(clean_votes)
    change_dtypes(df)
    df.reset_index(drop=True, inplace=True)
    print(df.info())
    df.to_csv(f'{save_name}_cleaned.csv', index=False)
    return df
```

```python
for file in glob('*_movies.csv'):
    df = pd.read_csv(file)
    df = clean_data(df, file.split('_')[0]+'_movies')
```

```python
all_data = pd.DataFrame()
for file in glob('*_movies_cleaned.csv'):
    df = pd.read_csv(file)
    all_data = pd.concat([all_data, df], axis=0, ignore_index=True)
all_data.drop('Unnamed: 0', axis=1, inplace=True)
all_data.drop_duplicates(subset=['movie', 'year', 'category', 'duration', 'rating', 'tagline', 'director', 'stars'], inplace=True, keep='last', ignore_index=True)
all_data.to_csv('all_movies_data.csv', index=False)
```
