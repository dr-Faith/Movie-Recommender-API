```python
from bs4 import BeautifulSoup
from selenium import webdriver
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy import Selector
import requests
import re, time, os
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
urls = [
    'https://en.wikipedia.org/wiki/List_of_South_Korean_dramas'
]

# constants 
KD_WIKI_URL = 'https://en.wikipedia.org/wiki/List_of_South_Korean_dramas'
```

```python
# Scraping Wikipedia for kdrama information and year

driver = webdriver.Chrome()

driver.get(KD_WIKI_URL)
drama_list = driver.find_elements(by='css selector', value='ul > li')
```

```python
def extract_movie_info():
    movie_info = [text for text in [info.text for info in drama_list] if re.match(r'^[\d\w\s]+\s\(\d{4}\)$', text)]
    return [(re.search(r'^[\d\s\w]+', text).group(0).strip(), re.search(r'\((\d+)\)$', text).group(1)) for text in movie_info]
```

```python
movies_info = extract_movie_info()    
```

```python
df = pd.DataFrame(movies_info)
df.to_csv('kdrama_year.csv', index=False)
```

```python
drama_texts = driver.find_elements(by='css selector', value='ul > li > i > a')
drama_info_links = [el.get_attribute('href') for el in drama_texts]
```

```python
with open('drama_links.text', 'a') as file:
    for line in drama_info_links:
        file.write(line + '\n')
```

```python
with open('drama_links.text', 'r') as file:
    links = file.readlines()
```

```python
len(links)
```

```python
# Using Scrapy Spider
movies_info = []

class KWikiScraper(scrapy.Spider):
    
    name = 'KWikiScraper'
    
    def start_requests(self):
        urls = links
        
        for url in urls:
            yield scrapy.Request(url=url, callback=self.add_movie_info)
            
    def add_movie_info(self, response):
        movie = response.css('#firstHeading > i::text').extract()
        attrs = response.css('.plainlist a::text').extract()
        movies_info.append((movie, attrs))
```

```python
process = CrawlerProcess()

process.crawl(KWikiScraper)
process.start()
```

```python
movies_df = pd.DataFrame(data={'movies': list(map(lambda x: str(x[0]).strip("['']"), movies_info)), 'attributes': list(map(lambda x: ', '.join(x[1]), movies_info))})
```

```python
class IMDBMovieScraper():
    """
        scrapes movie information by genres
    """
    IMDB_MOVIES_URL = 'https://www.imdb.com/feature/genre/?ref_=nv_ch_gr'
    NUM_PAGES = 100
    
    def __init__(self, genre, ):
        self.genre = genre
        self.movies_info = {
                'movie': [],
                'year': [],
                'category': [],
                'duration': [],
                'genres': [],
                'rating': [],
                'tagline': [],
                'director': [],
                'stars': [],
                'votes': []
            }
        self.movies_df = pd.DataFrame()
        
    def extract_elements(text):
        try:
            title = re.search(r'\d{1,2}\.\s([\w\W]+?)\s\(([\d\s–]+?)\)', text).group(1)
        except AttributeError:
            title = 'None'
        try:
            year = re.search(r'\d{1,2}\.\s([\w\W]+?)\s\(([\d\s–]+?)\)', text).group(2)
        except AttributeError:
            year = 'None'
        try:
            category = re.search(r'\([\d\s–]+?\)\n([\w\-]+)\s\|', text).group(1)
        except AttributeError:
            category = 'None'
        try:
            duration = re.search(r'\|\s(\d+)\smin\s\|', text).group(1)
        except AttributeError:
            duration = 'None'
        try:
            genres = re.search(r'\|\s\d+\smin\s\|\s([\w\W]+?)\n', text).group(1)
        except AttributeError:
            genres = 'None'
        try:
            rating = re.search(r'\|\s[\w\W]+?\n([\d\.]+)\s', text).group(1)
        except AttributeError:
            rating = 'None'
        try:
            tagline = re.search(r'\n[\w\W]+\n([\w\W]+)\.\n(Director|Stars)', text).group(1)
        except AttributeError:
            tagline = 'None'
        try:
            directors = re.search(r'Director:\s([\w\W]+)\s\|', text).group(1)
        except AttributeError:
            directors = 'None'
        try:
            stars = re.search(r'Stars:\s([\w\W]+)\n', text).group(1)
        except AttributeError:
            stars = 'None'
        try:
            votes = re.search(r'Votes:\s([\d\,]+)', text).group(1)
        except AttributeError:
            votes = 'None'

        return title, year, category, duration, genres, rating, tagline, directors, stars, votes
    
    def capitalize(word):
        return '-'.join(list(map(str.capitalize, word.split('-'))))
        
    def scrape(self):
        edge_driver = webdriver.Edge()
        edge_driver.get(IMDBMovieScraper.IMDB_MOVIES_URL)
        img = edge_driver.find_element(by='xpath', value=f'//img[@title="{IMDBMovieScraper.capitalize(self.genre)}"]')
        img.click()
        time.sleep(3)
        description = edge_driver.find_element(by='css selector', value='.desc > span')
        number_of_titles = int(description.text.split(' ')[-2].replace(',', ''))
        number_of_pages = round(number_of_titles / 50)
        
        # The actual number of pages is quite large, but it will take a longer time to scrape all those pages
        movie_item_parent_class = '//div[@class="lister-item-content"]'
        
        # Hence I will scrape for only 1000 pages
    
        # timing the process
        start_time = time.time()
        for page in range(IMDBMovieScraper.NUM_PAGES):
            movie_tiles = edge_driver.find_elements(by='xpath', value=movie_item_parent_class)
            next_button = edge_driver.find_element(by='xpath', value='//a[contains(@class, "next-page")]')

            for movie in movie_tiles:
                text = movie.text
                title, year, category, duration, genres, rating, tagline, director, stars, votes = IMDBMovieScraper.extract_elements(text)
                self.movies_info['movie'].append(title)
                self.movies_info['year'].append(year)
                self.movies_info['category'].append(category)
                self.movies_info['duration'].append(duration)
                self.movies_info['genres'].append(genres)
                self.movies_info['rating'].append(rating)
                self.movies_info['tagline'].append(tagline)
                self.movies_info['director'].append(director)
                self.movies_info['stars'].append(stars)
                self.movies_info['votes'].append(votes)

            if (page + 1) % 100 == 1 and page != 0:
                time_after_100 = time.time()
                print(f"Elapsed time = {(time_after_100 - start_time) / 60} minutes")
                print('*'*50)

            next_button.click()
            time.sleep(2)
            print(f'{page+1}/{IMDBMovieScraper.NUM_PAGES}')
        self.convert_to_csv()
        
    def convert_to_csv(self):
        self.movies_df = pd.DataFrame.from_dict(self.movies_info, orient='columns')
        self.movies_df.to_csv(f'{self.genre}_movies.csv', index=False)
        print("*****Process Completed*****")
```
