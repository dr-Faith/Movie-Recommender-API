```python
import requests
import numpy as np
import pandas as pd
import os, json, string
from dotenv import load_dotenv
```

```python
API_ENDPOINT_1 = 'https://api.themoviedb.org/3/trending/all/day'
API_ENDPOINT_2 = 'https://api.themoviedb.org/3/trending/all/week'
API_ENDPOINT_3 = 'https://api.themoviedb.org/3/discover/movie'
API_ENDPOINT_4 = 'https://api.themoviedb.org/3/discover/tv'
API_ENDPOINT_5 = 'https://api.themoviedb.org/3/movie/popular'
API_ENDPOINT_6 = 'https://api.themoviedb.org/3/tv/popular'

load_dotenv()

API_KEY = os.getenv('TMDB_API_KEY') 
```

```python
MOVIE_INFO_LIST = []
```

```python
def fetch_movies(num_pages, api_endpoint,  add_params=None):        
    
    params = {
        'api_key': API_KEY,
    }
    
    if add_params:
        for key, val in add_params:
            params[key] = val
       
    for page in range(1, num_pages+1):
        params['page'] = page
        response = requests.get(api_endpoint, params=params)
        results = response.json()['results']
        for json_data in results:
            try:
                MOVIE_INFO_LIST.append(
                    {
                        'backdrop_path': json_data['backdrop_path'] if 'backdrop_path' in json_data.keys() else np.nan,
                        'id': json_data['id'] if 'id' in json_data.keys() else np.nan,
                        'title': json_data['title'] if 'title' in json_data.keys() else json_data['name'],
                        'original_title': json_data['original_title'] if 'original_title' in json_data.keys() else json_data['original_name'],
                        'overview': json_data['overview'] if 'overview' in json_data.keys() else np.nan,
                        'poster_path': json_data['poster_path'] if 'poster_path' in json_data.keys() else np.nan,
                        'media_type': json_data['media_type'] if 'media_type' in json_data.keys() else np.nan,
                        'genre_ids': json_data['genre_ids'] if 'genre_ids' in json_data.keys() else np.nan,
                        'popularity': json_data['popularity'] if 'popularity' in json_data.keys() else np.nan,
                        'release_date': json_data['release_date'] if 'release_date' in json_data.keys() else json_data['first_air_date'],
                        'vote_average': json_data['vote_average'] if 'vote_average' in json_data.keys() else np.nan,
                        'vote_count': json_data['vote_count'] if 'vote_count' in json_data.keys() else np.nan
                    }
                )
            except:
                pass
        print(f"****page {page} done*****")
```

```python
df_ = pd.read_csv('tmdb_data/tmdb_movies.csv')
movie_ids = df_['id'].values
```

```python
### get reviews for each movie id
def get_reviews_by_batch(start_idx, end_idx):
    reviews = []
    for idx, id_ in enumerate(movie_ids[start_idx:end_idx]):
        response = requests.get(f'https://api.themoviedb.org/3/movie/{id_}/reviews', params={'api_key': API_KEY, 'page': 1})
        try:
            for review in response.json()['results']:
                reviews.append({
                    'movie_id': id_,
                    'username': review['author_details']['username'],
                    'rating': review['author_details']['rating']
                })
            print(f'***** movie {start_idx+idx+1} done *****')
        except:
            print(f'----- movie {start_idx+idx+1} error -----')
    with open('tmdb_data/reviews.txt', 'a') as file:
        for review in reviews:
            file.write(json.dumps(review, default=str) + '\n')
```

```python
get_reviews_by_batch(28000, 30438)
```
