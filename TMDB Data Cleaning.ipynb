```python
import pandas as pd
import numpy as np
import matplotlib

%matplotlib inline
```

```python
movies_df = pd.read_csv('tmdb_data/tmdb_movies.csv')
reviews_df = pd.read_csv('tmdb_data/reviews.csv')
```

```python
def user_id_gen(col):
    unique_users = reviews_df[col].unique()
    ids = {x: str(list(unique_users).index(x)) for x in unique_users}
    return ids
```

```python
reviews_df['username'] = reviews_df['username'].map(user_id_gen('username'))
```

```python
reviews_df.drop_duplicates(inplace=True)
```

```python
reviews_df.isna().sum()
```

```python
reviews_df.dropna(inplace=True)
```

```python
reviews_df.isna().sum()
```

```python
unique_mov_ids = reviews_df['movie_id'].unique()
```

```python
len(unique_mov_ids)
```

```python
len(movies_df['id'].unique())
```

```python
merged_df = pd.merge(left=movies_df, right=reviews_df, left_on='id', right_on='movie_id', how='inner')
```

```python
merged_df.shape
```

```python
merged_df.drop('movie_id', axis=1, inplace=True)
```

```python
merged_df_ = merged_df[['id', 'username', 'rating']]
```

```python
merged_df_.rename(columns={'id': 'movie_id', 'username': 'user_id'}, inplace=True)
```
